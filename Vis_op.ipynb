```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_scores(df,score_name):
    
    x = ['Ridge' , 'SVR' , 'AdaBoost' , 'RandomForest' , 'NeuralNet' , 'Stack_model']
    y = list(df.loc[:,score_name].values)
    #plt.grid(False)
    plt.figure(figsize = (15,10))
    plt.title("Comparison of {}".format(score_name))
    plt.bar(x,y,color = ['mediumpurple','dodgerblue','springgreen','lightsalmon' , 'lightslategrey' , 'm'])
    plt.xlabel("Models")
    plt.ylabel(score_name)
    plt.show()
```

```python
base_op = pd.read_csv('Base_learners_OP.csv')
```

```python
base_op
```

```python

#R2-Score
plot_scores(base_op , 'R2-Score')

#MSE 
plot_scores(base_op , 'MSE')
#Rmse
plot_scores(base_op , 'RMSE')
#MSE Log
plot_scores(base_op , 'MSE Log')
```
